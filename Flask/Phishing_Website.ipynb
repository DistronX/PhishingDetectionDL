```python
x = df.iloc[:,1:31].values
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout 
```

```python
Regression_Model = Sequential()
```

```python
Regression_Model.add(Dense(units=30, kernel_initializer="random_uniform", activation="relu"))
```

```python
Regression_Model.add(Dense(units=60, kernel_initializer="random_uniform", activation="relu"))
```

```python
Regression_Model.add(Dropout(0.2))
```

```python
Regression_Model.add(Dense(units=60, kernel_initializer="random_uniform", activation="relu"))
```

```python
Regression_Model.add(Dropout(0.2))
```

```python
Regression_Model.add(Dense(units=1, kernel_initializer="random_uniform", activation="relu"))
```

```python
Regression_Model.compile(optimizer = "adam",loss ="mean_squared_error", metrics = ["accuracy"] )
```

```python
x_train=np.asarray(x_train).astype(np.float32) 
y_train=np.asarray(y_train).astype(np.float32)
```

```python
history = Regression_Model.fit(x_train,y_train, batch_size =64,epochs = 500)
```

```python
ypred2 = Regression_Model.predict(x_test)
```

```python
y_pred3=[]
for i in ypred2:
    if i>0:
        y_pred3.append(1)
    else:
        y_pred3.append(-1)
```

```python
y_pred3
```

```python
from sklearn.metrics import accuracy_score
log=accuracy_score(y_test,y_pred3)
log
```
